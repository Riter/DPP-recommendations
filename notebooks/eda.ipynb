```python
# Заполнение пропущенных значений в столбце skills пустой строкой
df['skills'] = df['skills'].fillna('')

# Формирование датасета Team-Case
team_case_dataset = (
    df.groupby(["team_id", "team_title"])
    .agg({
        "topic_title": lambda x: list(x.unique()),  # Уникальные кейсы для команды
        "skills": lambda x: set(", ".join(x).split(", ")) if x.any() else set()  # Объединение всех навыков команды
    })
    .reset_index()
    .rename(columns={"topic_title": "case_titles", "skills": "team_skills"})
)
```

```python
team_case_dataset.head()
```

```python
case_team_dataset = (
    df.groupby(["topic_title", "topic_description"])
    .agg({
        "team_id": lambda x: list(x.unique()),  # Уникальные команды для кейса
        "skills": lambda x: set(", ".join(x).split(", ")) if x.any() else set()  # Все навыки участников команд
    })
    .reset_index()
    .rename(columns={"team_id": "team_ids", "skills": "required_skills"})
)
```
